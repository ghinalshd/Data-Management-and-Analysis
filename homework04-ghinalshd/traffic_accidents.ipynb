### 5. determine the top three streets(Use the ON STREET NAME column) that had the most accidents

For easier use and quicker implementation, I transformed the 'on street' column to a string type and replaced any 'nan' values to a string with an actual NaN (Not a Number) value, which is a standard representation of missing or null data in pandas. By doing so, we can keep the missing values, avoid dropping any rows, which in return, avoids any errors in our analysis.

```python
# Clean and transform the 'on street' column
data['on street'] = data['on street'].astype(str)
data['on street'] = data['on street'].replace('nan', np.nan)
# Since we have already changed the 'on street' column to a 'String' type, we can go ahead and directly calculate the street names with the most accidents
on_street_value_counts = data['on street'].value_counts()
# Print out the top 3 street addresses with the addresses
on_street_value_counts[0:3]
```

### 6. shows the number of accidents that occurred at each borough

For easier use and quicker implementation, I transformed the 'borough' column to a string type and replaced any 'nan' values to a string with an actual NaN (Not a Number) value, which is a standard representation of missing or null data in pandas. By doing so, we can keep the missing values, avoid dropping any rows, which in return, avoids any errors in our analysis.  

```python
# Clean and transform the 'borough' column
data['borough'].astype(str)
data['borough'].replace('nan', np.nan)
# List of given boroughs
boroughs = ["BRONX", "BROOKLYN", "QUEENS", "MANHATTAN"]
# Labeling all boroughs not in the list and the 'MISSING' values as 'OTHER'
data['borough'].apply(lambda x: x if x in boroughs else 'OTHER')
# Calculate the number of accidents in each borough and print out the values
borough_counts = data['borough'].value_counts()
borough_counts
```

```python
import matplotlib.pyplot as plt
# Plot the data
borough_counts.plot(kind='bar', figsize=(10,6), color='c')
plt.title('Number of Accidents Per Borough')
plt.xlabel('Borough')
plt.ylabel('Number of Accidents')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

### 8. what are the distributions of accidents based on the geo location (latitude & longitude)?

```python
df = data[data['longitude'] != 0] # Filters out rows where the longitude is 0 

import matplotlib.pyplot as plt
borough_colors = {
    'BRONX': 'red',
    'BROOKLYN': 'purple',
    'MANHATTAN': 'yellow',
    'QUEENS': 'blue',
    'STATEN ISLAND': 'turquoise'}

for borough, color in borough_colors.items():
    borough_df = df[df['borough'] == borough]
    plt.scatter(borough_df['longitude'], borough_df['latitude'], color=color, label=borough, alpha=0.7)
plt.xlabel('Latitude')
plt.ylabel('Longitude')
plt.legend()
plt.show()
```

I cleaned the data by filtering out all the rows where the longitude is equal to 0. Without doing so, the graph completely adjusted our graph where the '0' data points were relatively far from the other values. In reality, a '0' value could potentially be a mistakenly entered value or an incorrect one. By doing so, we were able to create a new dataframe with the updated data containing the geo-locations. As seen from the scatterplot above, the map resembles the one of New York with a few points in the incorrect borough, which could be a result of mislabelling or incorect entered values. 

### 9. shows the covariance between each pair of the columns

```python
# Define pairs for which covariance is to be computed
cov_pairs = [
    ('zip', 'number of persons injured'),
    ('number of persons injured', 'number of persons killed'),
    ('zip', 'number of cyclist injured'),
    ('zip', 'number of motorist injured')
]

# Iterate over pairs and compute/print covariance
for pair in cov_pairs:
    cov_value = data[pair[0]].cov(data[pair[1]])
    print(f'Covariance between {pair[0]} and {pair[1]}: {cov_value:.2f}')
```

I have calculated the covariance between zip codes and number of persons injured, number of cyclist injured and number of motorist injured as well as the covariance between number of persons injured and killed.

The reason why I used the 'zip' column instead of the 'borough' column is due to the fact that covariance is calculated using numerical values only. The 'borough' column contains categorical data with string values, hence calculating the covariance may not be meaningful. Using the 'zip' column means that we are able to identify the borough from the zip codes by using the following list:
* Manhattan: 10001-10282
* Staten Island: 10301-10314
* Bronx: 10451-10475
* Queens: 11004-11109, 11351-11697
* Brooklyn: 11201-11256 

Observations and Interpretations: 
1) Covariance between zip and number of persons injured: 4.75

This positive value suggests that as the zip code value increases (areas closer to Queens and Brooklyn), the number of persons injured tends to increase as well. 

2) Covariance between number of persons injured and number of persons killed: 0.00

A covariance close to 0 indicates that there is not a strong linear relationship between the number of persons injured and killed in an accident. It suggests that accidents with more injuries don't necessarily have more fatalities.

3) Covariance between zip and number of cyclist injured: -4.26

This negative value suggests that as the zip code value increases (areas closer to Queens and Brooklyn), the number of cyclist injuries tends to decrease. 

4) Covariance between zip and number of motorist injured: 10.11

A positive value implies that as the zip code value increases (areas closer to Queens and Brooklyn), the number of motorist injuries tends to increase as well.
